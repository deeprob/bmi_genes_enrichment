```python
import os
import pandas as pd
```

# External gene set lipid

```python
groups = ["white_british", "white_british_male", "white_british_female", "post_menopause", "pre_menopause"]
phenos = ["high_bmi", "low_bmi"]
store_dir = "/data5/deepro/ukbiobank/analysis/bmi_genes_enrichment/data/external_enrich_results"
```

```python
dfs = []

for pheno in phenos:
    for group in groups:
        enrichment_results = os.path.join(store_dir, "lipid", pheno, group, "external_enrich.csv")
        df_enrich = pd.read_csv(enrichment_results)
        df_enrich["phenos"] = pheno
        df_enrich["groups"] = group
        dfs.append(df_enrich)
            
lipid_df = pd.concat(dfs, axis=0)
```
